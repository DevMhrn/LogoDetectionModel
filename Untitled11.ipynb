```python
import os

# Path to your train and valid directories containing labels
train_dir = '/content/Pepsi-Logo-Detection-1/train'
valid_dir = '/content/Pepsi-Logo-Detection-1/valid'

def update_labels(directory):
    # Iterate through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):  # Assuming labels are stored as text files
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r+') as file:
                # Read the content of the label file
                content = file.read().strip()
                # Split the content by spaces
                parts = content.split()
                if parts[0] == '0':  # Check if the class name is '0'
                    parts[0] = '1'  # Change class name to '1'
                    # Join the parts back into a string
                    new_content = ' '.join(parts)
                    # Write the updated content back to the file
                    file.seek(0)
                    file.write(new_content)
                    file.truncate()

# Update labels in both train and valid directories
update_labels(train_dir)
update_labels(valid_dir)
```

```python
#As both datasets downloaded as seperate dataset setting one dataset with one class as 1
import os
def change_class_id(directory, old_id=0, new_id=1):
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                lines = file.readlines()

            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts[0] == str(old_id):
                    parts[0] = str(new_id)
                new_lines.append(' '.join(parts))

            with open(filepath, 'w') as file:
                file.write('\n'.join(new_lines))
            print(f'Updated {filename}')


label_files_directory = '/content/Pepsi-Logo-Detection-1/train/labels'
change_class_id(label_files_directory)
```

```python
import os
import shutil

def move_files(source_dir, dest_dir, num_files):
    # Create the destination directory if it doesn't exist
    os.makedirs(dest_dir, exist_ok=True)
    os.makedirs(os.path.join(dest_dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(dest_dir, 'labels'), exist_ok=True)

    # Get list of files in the source directory
    files = os.listdir(os.path.join(source_dir, 'images'))

    # Move the first num_files files to the destination directory
    for i in range(num_files):
        # Move image file
        image_file = os.path.join(source_dir, 'images', files[i])
        shutil.move(image_file, os.path.join(dest_dir, 'images', files[i]))

        # Move label file
        label_file = os.path.join(source_dir, 'labels', files[i].replace('.jpg', '.txt'))  # Assuming label file extension is '.txt'
        shutil.move(label_file, os.path.join(dest_dir, 'labels', files[i].replace('.jpg', '.txt')))

        print(f'Moved {files[i]} to {dest_dir}')

    print(f'\nSuccessfully moved {num_files} files from {source_dir} to {dest_dir}')

# Define paths
train_dir = '/content/Pepsi-Logo-Detection-1/train'
test_dir = os.path.join(train_dir, 'test')

# Move first 50 files from train directory to test directory
move_files(train_dir, test_dir, 50)
```
